```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
```

```python
vaccine_misinformation_df = pd.read_excel (r'tweets.xlsx') 
vaccine_misinformation_df
```

```python
twit_df = vaccine_misinformation_df['Tweets']
```

# Text Preprocessing
## Text Segmentation

```python
segments = []
for x in twit_df:
    segments.append(nltk.tokenize.sent_tokenize(x))
segments
```

## Case Folding
### To lower

```python
results = []
for x in segments:
    segment = []
    for y in x:
        segment.append(y.lower())
    results.append(segment)
results
```

### remove numeric

```python
import re
```

```python
for x in results:
    for y in x:
        results[results.index(x)][x.index(y)] = re.sub(r"\d", "", y)
results
```

### Remove punctuation

```python
import string
for x in results:
    for y in x:
        results[results.index(x)][x.index(y)] = y.translate(str.maketrans("","",string.punctuation))
results
```

### Stopword removal

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
tokens = []
for x in results:
    token = []
    for y in x:
        twit = word_tokenize(y)
        token.append(twit)
    tokens.append(token)
tokens
```

```python
listStopword = set(stopwords.words('english'))
```

```python
tokens_clear = []
for x in tokens:
    token = []
    for y in x:
        removed = []
        for z in y:
            if z not in listStopword:
                removed.append(z)
        token.append(removed)
    tokens_clear.append(token)
tokens_clear
```

## Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for x in tokens_clear:
    roots_en = []
    for y in x:
        root_en = []
        for z in y:
            root_en.append(ps.stem(z))
        roots_en.append(root_en)
    tokens_clear[tokens_clear.index(x)]=(roots_en)
tokens_clear
```

```python
fix_token_arr = []
for x in tokens_clear:
    token_in_sentence = []
    for y in x:
        for z in y:
            token_in_sentence.append(z)
    fix_token_arr.append(token_in_sentence)
fix_token_arr
```

```python
fix_token_arr
```

```python
vaccine_misinformation_df['preprocessed_tweet'] = fix_token_arr
```

```python
vaccine_misinformation_df
```

# TF-IDF

## Term Frequency (TF)

```python
def calc_TF(document):
    # Counts the number of times the word appears in review
    TF_dict = {}
    for term in document:
        if term in TF_dict:
            TF_dict[term] += 1
        else:
            TF_dict[term] = 1
    # Computes tf for each word
    for term in TF_dict:
        TF_dict[term] = TF_dict[term] / len(document)
    return TF_dict

vaccine_misinformation_df["TF_dict"] = vaccine_misinformation_df['preprocessed_tweet'].apply(calc_TF)

vaccine_misinformation_df["TF_dict"].head()
```

```python
# Check TF result
index = 90

print('%20s' % "term", "\t", "TF\n")
for key in vaccine_misinformation_df["TF_dict"][index]:
    print('%20s' % key, "\t", vaccine_misinformation_df["TF_dict"][index][key])
```

## Inverse Document Frequency (IDF)

```python
def calc_DF(tfDict):
    count_DF = {}
    # Run through each document's tf dictionary and increment countDict's (term, doc) pair
    for document in tfDict:
        for term in document:
            if term in count_DF:
                count_DF[term] += 1
            else:
                count_DF[term] = 1
    return count_DF

DF = calc_DF(vaccine_misinformation_df["preprocessed_tweet"])
DF
```

```python
n_document = len(vaccine_misinformation_df)

def calc_IDF(__n_document, __DF):
    IDF_Dict = {}
    for term in __DF:
        IDF_Dict[term] = np.log(__n_document / (__DF[term] + 1))
    return IDF_Dict
  
#Stores the idf dictionary
IDF = calc_IDF(n_document, DF)
IDF
```

## TF-IDF

```python
# calc TF-IDF
def calc_TF_IDF(TF):
    TF_IDF_Dict = {}
    #For each word in the review, we multiply its tf and its idf.
    for key in TF:
        TF_IDF_Dict[key] = TF[key] * IDF[key]
    return TF_IDF_Dict

#Stores the TF-IDF Series
vaccine_misinformation_df["TF-IDF_dict"] = vaccine_misinformation_df["TF_dict"].apply(calc_TF_IDF)
```

```python
# Check TF-IDF result
index = 90

print('%20s' % "term", "\t", '%10s' % "TF", "\t", '%20s' % "TF-IDF\n")
for key in vaccine_misinformation_df["TF-IDF_dict"][index]:
    print('%20s' % key, "\t", vaccine_misinformation_df["TF_dict"][index][key] ,"\t" , vaccine_misinformation_df["TF-IDF_dict"][index][key])
```

```python
# sort descending by value for DF dictionary 
sorted_DF = sorted(DF.items(), key=lambda kv: kv[1], reverse=True)[:50]

# Create a list of unique words from sorted dictionay `sorted_DF`
unique_term = [item[0] for item in sorted_DF]

def calc_TF_IDF_Vec(__TF_IDF_Dict):
    TF_IDF_vector = [0.0] * len(unique_term)

    # For each unique word, if it is in the review, store its TF-IDF value.
    for i, term in enumerate(unique_term):
        if term in __TF_IDF_Dict:
            TF_IDF_vector[i] = __TF_IDF_Dict[term]
    return TF_IDF_vector

vaccine_misinformation_df["TF_IDF_Vec"] = vaccine_misinformation_df["TF-IDF_dict"].apply(calc_TF_IDF_Vec)

print("print first row matrix TF_IDF_Vec Series\n")
print(vaccine_misinformation_df["TF_IDF_Vec"][0])

print("\nmatrix size : ", len(vaccine_misinformation_df["TF_IDF_Vec"][0]))
```

```python
# Convert Series to List
TF_IDF_Vec_List = np.array(vaccine_misinformation_df["TF_IDF_Vec"].to_list())

# Sum element vector in axis=0 
sums = TF_IDF_Vec_List.sum(axis=0)

data = []

for col, term in enumerate(unique_term):
    data.append((term, sums[col]))
    
ranking = pd.DataFrame(data, columns=['term', 'rank'])
ranking.sort_values('rank', ascending=False)
```
